```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
```

```python
load_dotenv()


model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
prompt1=PromptTemplate(template="""Generate a detailed case study on {topic} """,
                        input_variables=['topic'])

                    
prompt2=PromptTemplate(template="""Summarize this into small paras for each of the topics/heading provided in the provided {text} """,
                        input_variables=['text'])


parser=StrOutputParser()

chain=prompt1|model|parser|prompt2|model|parser


result=chain.invoke({"topic":"Time Space Singularity"})


print(prompt1.content)

print(result)



```
