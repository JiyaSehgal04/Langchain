```python
from langchain_core.documents import Document
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled
import faiss
import langchain_huggingface
from langchain_google_gemini import ChatGoogleGenerativeAI
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from dotenv import load_dotenv

```

```python
load_dotenv()


#1. Indexing
video_id="DlIAd4Rtkr8"
try:
    transcript_list = YouTubeTranscriptApi.get_transcript(video_id, languages=['en'])
    
    transcript = " ".join(chunk['text'] for chunk in transcript_list)
except TranscriptsDisabled:
    print("Transcripts are disabled for this video")
    
```

```python
#2.Indexing: Text Splitting
splitter=RecursiveCharacterTextSplitter(chunk_size=300,chunk_overlap=5)
chunks=splitter.create_documents([transcript])
```

```python
#3.Embeding creation
embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")


vector_store=faiss.from_documents(chunks,embedding=embeddings)

```

```python
#Retriever
retriever=vector_store.as_retriever(search_type="similarity",searc_kwargs={"k":2})
```

```python

prompt=PromptTemplate(
    template="",
    input_variables=['context','question']
)


question="What is the main topic of the video?"
retrived_documents=retriever.invoke(question)
context_text="\n\n".join([doc.page_content for doc in retrived_documents])


final_prompt=promt.invoke({"context":context_text,"question":question})
model=ChatGoogleGenerativeAI(model_name="gemini-2.5-flash")
response=model.invoke(final_prompt)

print(response.content)


```
