```python
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
from dotenv import load_dotenv
# from langchain_core.retrievers import VectorStoreRetriever
```

```python
load_dotenv()


documents=[
    Document(
    page_content="Iceland is one of the most volcanically active regions in the world due to its location on the Mid-Atlantic Ridge, where the North American and Eurasian tectonic plates are moving apart. This geological setting allows magma from the Earth’s mantle to rise close to the surface, resulting in frequent volcanic eruptions. Iceland has over thirty active volcanic systems, and eruptions can vary from slow lava flows to explosive ash-producing events, depending on magma composition and interaction with ice or water."
),
Document(
    page_content="Many Icelandic volcanoes lie beneath glaciers, creating unique eruption styles known as subglacial eruptions. When magma erupts beneath thick ice, it rapidly melts the ice and produces large volumes of meltwater, which can lead to sudden and destructive floods called jökulhlaups. These eruptions also generate fine volcanic ash as magma interacts with meltwater, increasing the explosiveness of the eruption."
),
Document(
    page_content="The 2010 eruption of the Eyjafjallajökull volcano in Iceland highlighted the global impact of volcanic activity in ice-covered regions. The eruption occurred beneath a glacier, causing significant ash production that disrupted air travel across Europe for several days. The fine ash particles posed a serious risk to aircraft engines, demonstrating how Icelandic volcanic eruptions can affect regions far beyond the island itself."
),
Document(
    page_content="Volcanic eruptions in Iceland can influence the environment by releasing gases such as sulfur dioxide into the atmosphere. These gases can form aerosols that temporarily cool the climate by reflecting sunlight. Additionally, ash deposition from eruptions can affect soil fertility, water quality, and nearby ecosystems, particularly in areas where volcanic ash settles on glaciers or is transported by wind."
),
Document(
    page_content="Iceland closely monitors its volcanic systems using seismic sensors, GPS measurements, and satellite observations to detect early signs of eruptions. This monitoring is especially important for volcanoes located beneath glaciers, as eruptions can trigger sudden floods and widespread ash clouds. Early warning systems help authorities reduce risks to local communities and international transportation."
)
]
embeddings=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
)
vector_store=Chroma.from_documents(
    documents=documents,
    embedding=embeddings,
    collection_name="volcanic-eruptions"
)

retriever=vector_store.as_retriever()

query="What is the most recent volcanic eruption in Iceland?"
docs=retriever.invoke(query)


print(docs)
```
