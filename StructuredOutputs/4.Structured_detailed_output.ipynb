```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from typing import TypedDict, Optional,List, Annotated
```

```python
load_dotenv()

model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")


#Schema
class Review(TypedDict):
    key_themes:Annotated[List[str],"The key themes of the text /n"]
    summary:Annotated[str,"A concise summary of the Input paragraph /n"]  
    sentiment:Annotated[str,"The sentiment of the Input paragraph, either be positive or negative"]
    pros:Annotated[List[str],"The pros of the Input paragraph"]
    cons:Annotated[List[str],"The cons of the Input paragraph"]


structured_model=model.with_structured_output(Review)

prompt="""  
In the final moments of the story, Mr. Darcy openly declares his love for Elizabeth with humility and sincerity, admitting that his earlier pride and reserved manner concealed feelings that had grown stronger despite her rejection. He acknowledges his past mistakes, particularly his interference in her sister’s happiness, and explains that his actions were driven by concern rather than arrogance. Elizabeth, having reassessed Darcy’s character and witnessed his quiet acts of kindness and personal growth, recognizes that her own prejudice had blinded her to his true nature. Their mutual confessions reflect emotional maturity on both sides, as pride gives way to understanding and prejudice is replaced by respect, ultimately uniting them in a relationship founded on honesty, equality, and genuine affection.
"""


result=structured_model.invoke(prompt)



print(result)
```
