```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from typing import TypedDict, Annotated, Optional, Literal
from pydantic import BaseModel, Field
```

```python
#Specify the env file and the model
load_dotenv()
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
#Schema
class Review(BaseModel):
    key_themes:list[str]=Field(description="Write down 3 themes discussed in the review in a list.")
    summary:str=Field(description="Write a concise summary of the review in 2-3 sentences.")
    sentiment:Literal["positive","negative"]=Field(description="What is the sentiment of the review, either be positive or negative")
    # pros:list[str]=Field(description="Write down the pros of the review in a list")
    # cons:list[str]=Field(description="Write down the cons of the review in a list")
    name:str=Field(description="What is the name of the reviewer?")
    
    
st_model=model.with_structured_output(Review)
    
prompt="""  In the final moments of the story, Mr. Darcy openly declares his love for Elizabeth with humility and sincerity, admitting that his earlier pride and reserved manner concealed feelings that had grown stronger despite her rejection. He acknowledges his past mistakes, particularly his interference in her sister’s happiness, and explains that his actions were driven by concern rather than arrogance. Elizabeth, having reassessed Darcy’s character and witnessed his quiet acts of kindness and personal growth, recognizes that her own prejudice had blinded her to his true nature. Their mutual confessions reflect emotional maturity on both sides, as pride gives way to understanding and prejudice is replaced by respect, ultimately uniting them in a relationship founded on honesty, equality, and genuine affection.
"""

result=st_model.invoke(prompt)
print(result)
    
```
