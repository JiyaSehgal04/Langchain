```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel, RunnableSequence
```

```python
load_dotenv()

promp1=PromptTemplate(template="Give a linkedin Post about {topic}",
                      input_variables=["topic"])

promp2=PromptTemplate(template="Give an instagram description about {topic}",
                      input_variables=["topic"])

model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
parser=StrOutputParser()

parallel_chain=RunnableParallel({
    "tweet":RunnableSequence(promp1,model,parser),
    "instagram":RunnableSequence(promp2,model,parser)
})

result=parallel_chain.invoke({"topic":"Effects on  animals due to rising temperatures in Antarctica"})

print(result)
```
