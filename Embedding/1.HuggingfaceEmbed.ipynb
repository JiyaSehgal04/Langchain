```python
from sklearn.metrics.pairwise import cosine_similarity
from dotenv import load_dotenv
import os
from langchain_huggingface import HuggingfaceEmbeddings
```

Embeddings turn text (or images, audio, etc.) into numbers that capture meaning, so computers can compare, search, and reason about similarity. They are used for:
1. Semantic Similarity: Providing the text which mean the same thing
2. Search the keyword whose meaning is same
3. Using for Cosine Similarity: Vectors pointing in the same direction mean the same thing while the ones in the opposite direction


```python
load_dotenv()
#we use embeddings for converting text into vectors to capture meaning and not the exact or same words
#we use semantic or vector search so that the search is not based on the exact words but the meaning
embeddings=HuggingfaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")


text="New York is a very expensive place to live"


Documents=[
    "I love New York",
    "Pizza is a great meal",
    "Paris is the capital of France"
]

#Converting the provided text to embeddings
result=embeddings.embed_query(text)



result_doc=embeddings.embed_documents(Documents)
similarity_score=cosine_similarity([result],result_doc)
print(similarity_score)
```
