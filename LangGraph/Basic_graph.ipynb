```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Dict
from IPython.display import Image, display
```

```python
class messageState(TypedDict):
    message:str

def greetingnode(state:messageState) -> messageState:
    """Simple node that adds a greeting message to the state """
    state['message']="Hello " + state['message'] + "Welcome to Jiya's realm"
    return state


graph=StateGraph(messageState)
graph.add_node('initial',greetingnode)
graph.add_edge(START, 'initial')
graph.add_edge('initial', END)


bot=graph.compile()
graph_png=bot.get_graph().draw_mermaid_png()
display(Image(graph_png))
```

```python
response=bot.invoke({"message":"Paul, "})
print(response)
```
