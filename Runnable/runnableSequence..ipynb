```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence
```

```python
load_dotenv()

prompt=PromptTemplate(template="Tell me a joke about {topic}",
                      input_variables=["topic"])

parser=StrOutputParser()

prompt_2=PromptTemplate(template="Give me a summary of this text: {text}",
                        input_variables=["text"])
                
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")

runnable=RunnableSequence(prompt,model,parser,prompt_2,model,parser)


result=runnable.invoke({"topic":"Penguins"})

print(result)

```
