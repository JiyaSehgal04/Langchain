```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    TokenTextSplitter
)
```

```python
text=""" Rising temperatures in Antarctica are having profound and cascading effects on its wildlife and marine ecosystems, primarily driven by the rapid loss of sea ice and warming ocean waters. Sea ice serves as a critical habitat and breeding platform for species such as emperor and Adélie penguins, whose populations are declining as reduced ice coverage disrupts nesting, chick survival, and access to food. Warmer ocean temperatures are also altering the distribution and abundance of Antarctic krill, a keystone species that forms the foundation of the Southern Ocean food web, directly impacting seals, whales, and seabirds that depend on krill for survival. Additionally, melting glaciers increase freshwater input into the ocean, changing salinity levels and nutrient mixing, which affects phytoplankton productivity and, in turn, higher trophic levels. Antarctic seals, including Weddell and crabeater seals, face increased stress due to habitat loss, shifting prey availability, and greater exposure to predators and human activity. Collectively, these changes threaten the long-term stability of Antarctic biodiversity and highlight the sensitivity of polar ecosystems to even small increases in global temperatures.
"""


splitter=RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=10
)


result=splitter.split_text(text)

print(result)



```
